## 6. Context Manager

ë¦¬ì†ŒìŠ¤ë¥¼ ìë™ìœ¼ë¡œ ì •ë¦¬í•´ì£¼ëŠ” êµ¬ë¬¸/íŒ¨í„´. ì¦‰, ìë™í™” ê³¼ì •.

```python
from contextlib import contextmanager

@contextmanager # ë°ì½”ë ˆì´í„°(decorator): 'í•¨ìˆ˜'ë‚˜ 'í´ë˜ìŠ¤'ë¥¼ ê¾¸ë©°ì£¼ëŠ” ì—­í• ì´ë¼ê³  í•¨. ì¦‰, contextmanagerê°€ ê¸°ëŠ¥(ë‚´ê°€ ë§Œë“  í•¨ìˆ˜/í´ë˜ìŠ¤)ì„ ì¶”ê°€í•´ì¤Œ.
def my_connection():

    # ì—°ê²° ì—´ê¸°
    conn = pymysql.connect(
        host='127.0.0.1',
        user='root',
        password='1234',
        database='test_pipeline'
    )

    try:
        yield conn # yieldì—ì„œ ì ì‹œ ë©ˆì¶”ê³ , connì„ withì— ë„˜ê¹€. 'ì–‘ë³´í•˜ë‹¤'ì˜ yield ëœ»ì¸ê°€ ë´„.

    except Exception as e:
        logging.error(f"Error: {e}")
        raise

    finally:
        conn.close()

with my_connection() as conn: # connì„ ë°›ì•„ì„œ ì‹¤ì œ ì‘ì—…ì„ í•˜ëŠ” ê³³
    cursor = conn.cursor()
    cursor.execute("select count(*) from sales")
    result = cursor.fetchone()
    print(result)
```
